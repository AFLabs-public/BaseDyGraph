# Baseline Ladder — no-graph vs static vs dynamic, across seeds

Produces the central table of the thesis: does a **dynamic** latent graph beat a
**static** one beat **no graph** at next-state prediction, and does only the
dynamic model recover the regime graph?

For each `spatial_module_type in {none, static_graph, dynamic_graph}` and each
seed, this trains a model on a **fixed** synthetic dataset (so variance is from
model init / training only), evaluates it, and tabulates
`lift_over_persistence`, `graph_corr`, and `graph_auroc` with mean ± std.

**Run after** the cells that define `ModelConfig`,
`DiscreteSTGraphLightningModule`, `DiscreteStateDataModule`,
`SyntheticDiscreteGraphGenerator` / `SyntheticGraphDataConfig`, and the
`full_evaluation_report` from the evaluation-utilities notebook.

```python
def build_datamodule(
    synth_cfg: "SyntheticGraphDataConfig",
    splits: Dict[str, int],
    batch_size: int = 64,
) -> Tuple["DiscreteStateDataModule", torch.Tensor]:
    """Generate synthetic data once and wrap it in a DataModule.

    Returns (datamodule, regime_graphs). regime_graphs is the ground-truth
    (R, N, N) tensor used for graph-recovery evaluation.
    """
    gen = build_generator(synth_cfg, lead_lag=False)  # ContemporaneousGraphGenerator
    data = gen.generate()
    state_ids, regimes = data["state_ids"], data["regimes"]

    n = state_ids.size(0)
    assert splits["train"] + splits["val"] + splits["test"] == n, "splits must sum to num_samples"
    a, b = splits["train"], splits["train"] + splits["val"]

    dm = DiscreteStateDataModule(
        train_tensor=state_ids[:a], val_tensor=state_ids[a:b], test_tensor=state_ids[b:],
        train_regimes=regimes[:a], val_regimes=regimes[a:b], test_regimes=regimes[b:],
        batch_size=batch_size, num_workers=0,
    )
    return dm, data["regime_graphs"]
```

## 2. Train a single config

`pl.seed_everything` makes each run reproducible. Logging / checkpointing /
progress bars are off so the ladder loop stays quiet and fast.


```python
def train_one_config(
    spatial_module_type: str,
    seed: int,
    base_cfg_kwargs: Dict,
    dm: "DiscreteStateDataModule",
    true_regime_graphs: torch.Tensor,
    num_states: int,
    max_epochs: int = 30,
    lr: float = 1e-3,
    progress_bar: bool = True,
    use_wandb: bool = False,
    wandb_project: str = "BaseGraph Ablation",
    wandb_group: Optional[str] = None,
) -> Dict[str, float]:
    """Train one (config, seed) and return its evaluation report.

    If use_wandb, logs to W&B with run name f"{spatial_module_type}_seed{seed}"
    under `wandb_project`, grouped by `wandb_group` so the rungs compare cleanly.
    """
    pl.seed_everything(seed, workers=True)

    cfg = ModelConfig(**base_cfg_kwargs, spatial_module_type=spatial_module_type)
    model = DiscreteSTGraphLightningModule(cfg, lr=lr, true_regime_graphs=true_regime_graphs)

    logger = False
    if use_wandb:
        from lightning.pytorch.loggers import WandbLogger
        logger = WandbLogger(
            project=wandb_project,
            name=f"{spatial_module_type}_seed{seed}",
            group=wandb_group,
            config={**base_cfg_kwargs, "spatial_module_type": spatial_module_type,
                    "seed": seed, "max_epochs": max_epochs, "lr": lr},
        )

    trainer = pl.Trainer(
        max_epochs=max_epochs,
        accelerator="auto", devices="auto",
        logger=logger, enable_checkpointing=False, enable_progress_bar=progress_bar,
        num_sanity_val_steps=0, deterministic=False,
        log_every_n_steps=10,   # small dataset (~43 steps/epoch); 50 would log <1x/epoch
    )
    trainer.fit(model, datamodule=dm)

    report = full_evaluation_report(
        model, dm.val_dataloader(), num_states=num_states,
        true_regime_graphs=true_regime_graphs,
    )
    report["spatial_module_type"] = spatial_module_type
    report["seed"] = seed

    if use_wandb:
        import wandb
        # log the final summary metrics so they're visible on the run overview too
        wandb.log({f"final/{k}": v for k, v in report.items()
                   if isinstance(v, (int, float))})
        wandb.finish()

    return report

```

## 3. Run the full ladder

Three configs × the seeds you pass. `base_cfg_kwargs` must **not** contain
`spatial_module_type` (it is set per rung).


```python

def run_ladder(
    seeds: List[int],
    dm: "DiscreteStateDataModule",
    true_regime_graphs: torch.Tensor,
    num_states: int,
    base_cfg_kwargs: Dict,
    max_epochs: int = 30,
    lr: float = 1e-3,
    progress_bar: bool = True,
    use_wandb: bool = False,
    wandb_project: str = "BaseGraph Ablation",
    wandb_group: Optional[str] = None,
) -> List[Dict[str, float]]:
    rows = []
    for stype in LADDER:
        for seed in seeds:
            report = train_one_config(
                stype, seed, base_cfg_kwargs, dm, true_regime_graphs,
                num_states, max_epochs=max_epochs, lr=lr, progress_bar=progress_bar,
                use_wandb=use_wandb, wandb_project=wandb_project, wandb_group=wandb_group,
            )
            rows.append(report)
            corr = report.get("graph_corr", float("nan"))
            print(f"[{stype:<14} seed {seed}] "
                  f"acc={report['model_acc']:.4f}  "
                  f"lift={report['lift_over_persistence']:+.4f}  "
                  f"corr={corr:.3f}")
    return rows
```

## 4. Aggregate and compare

Mean ± std per rung, plus a **paired** comparison across seeds (dynamic − static
and dynamic − none) — pairing by seed removes seed-to-seed noise and is the
right way to ask "does dynamic help" with few runs.


```python
def _mean_std(values: np.ndarray) -> Tuple[float, float]:
    """NaN-safe mean/std; returns (nan, nan) if everything is NaN."""
    v = values[~np.isnan(values)]
    return (float(v.mean()), float(v.std())) if v.size else (float("nan"), float("nan"))


def summarize_ladder(rows: List[Dict[str, float]]) -> Dict[str, Dict[str, Tuple[float, float]]]:
    metrics = ["model_acc", "lift_over_persistence", "graph_corr", "deviation_corr"]
    summary = {}
    for stype in LADDER:
        vals = {
            m: np.array([r.get(m, np.nan) for r in rows if r["spatial_module_type"] == stype], float)
            for m in metrics
        }
        summary[stype] = {m: _mean_std(v) for m, v in vals.items()}

    header = f"{'config':<15}" + "".join(f"{m:>22}" for m in metrics)
    print(header); print("-" * len(header))
    for stype in LADDER:
        line = f"{stype:<15}"
        for m in metrics:
            mean, std = summary[stype][m]
            line += f"{mean:>11.4f} ± {std:<8.4f}"
        print(line)
    return summary


def paired_comparison(rows: List[Dict[str, float]], a: str, b: str, metric: str = "lift_over_persistence"):
    """Paired (by seed) difference a - b on a metric. Reports mean diff and,
    if scipy is available, a paired t-test p-value."""
    by_seed = lambda s: {r["seed"]: r.get(metric, np.nan) for r in rows if r["spatial_module_type"] == s}
    da, db = by_seed(a), by_seed(b)
    seeds = sorted(set(da) & set(db))
    diffs = np.array([da[s] - db[s] for s in seeds], float)
    out = {"n_pairs": len(seeds), "mean_diff": float(np.nanmean(diffs))}
    try:
        from scipy import stats
        t, p = stats.ttest_rel(
            [da[s] for s in seeds], [db[s] for s in seeds]
        )
        out["t_stat"], out["p_value"] = float(t), float(p)
    except Exception:
        out["note"] = "install scipy for a paired t-test; with few seeds report mean_diff ± spread"
    print(f"{a} - {b} on {metric}: {out}")
    return out
```

## 5. Run it


```python
# uses the globals from the config cell above
dm, regime_graphs = build_datamodule(synth_cfg, SPLITS, batch_size=BATCH_SIZE)
dm.setup()

rows = run_ladder(SEEDS, dm, regime_graphs, num_states=NUM_STATES,
                  base_cfg_kwargs=BASE_CFG_KWARGS, max_epochs=MAX_EPOCHS, lr=LR,
                  progress_bar=PROGRESS_BAR,
                  use_wandb=USE_WANDB, wandb_project=WANDB_PROJECT, wandb_group=WANDB_GROUP)

summarize_ladder(rows)
paired_comparison(rows, "dynamic_graph", "static_graph", "lift_over_persistence")
paired_comparison(rows, "dynamic_graph", "none", "lift_over_persistence")
```

**How to read the table.**

- `lift_over_persistence` rising `none -> static -> dynamic` is the core claim.
  If static already matches dynamic on lift, the *dynamics* aren't helping
  prediction even if they help recovery — report that honestly.
- `graph_corr` / `graph_auroc` should be near-chance for `none` (no graph) and
  `static_graph` (can't track regimes) and clearly higher for `dynamic_graph`.
  That contrast is the recovery story.
- A clean **negative** result (dynamic doesn't beat static) is still a valid
  thesis finding — the paired comparison is what makes either conclusion
  defensible. Three seeds is a minimum; use 5+ before claiming significance.
